# 2.Line Plot 

```python
df_imm= pd.read_csv('immigrants_emigrants_by_age.csv')
df_imm.head()
```

```python
table1 = pd.pivot_table(data = df_imm, index = 'Year', values = 'Immigrants', aggfunc = np.sum)
table1
```

```python
table2 = pd.pivot_table(data = df_imm, index = 'Year', values = 'Emigrants', aggfunc = np.sum)
table2
```

```python
plt.figure(figsize=(10,8))
plt.style.use('seaborn-dark')
plt.plot(table1.index, table1['Immigrants'],color='r')
plt.plot(table2.index, table2['Emigrants'],color='g')
plt.legend(["Immigrants", "Emigrants"], loc ="center left",fontsize=22)
plt.title('Immigrants and Emigrants in different Years(Line Plot)',fontsize=25)
plt.xlim(left=2013,right=2018)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number(Immigrants/Emigrants)',fontsize=20)
plt.show()
```

# 3.Scatter Plot

```python
df_pop= pd.read_csv('population.csv')
df_pop.head()
```

```python
tablef = pd.pivot_table(data = df_pop, index = 'district_code', values = 'number', aggfunc = np.sum)
tablef
```

```python
#scatter plots 
plt.figure(figsize=(16,6))
plt.style.use('seaborn')
plt.scatter(tablef.index,tablef['number'],color='red') 
plt.xlabel('District Code',fontsize=20) 
plt.ylabel('Number of population(in Million)',fontsize=20) 
plt.title('Population in Different Districts',fontsize=25) 
plt.show()
```
